```python
# re.findall()is used to find same element from list and it back to the function 

# re.findall(r'\d+', 'hello 42 I\'m a 32 string 30')  # ['42', '32', '30']

import re


txt = "The rain in Spain, i love rain"
x = re.findall('rain', txt)
print(x)
```

```python
len(x)
```

```python
#  \d to match the number in a string output formate ['5', '0']
txt = "The rain in Spain, i love rain 50"
x = re.findall('\d', txt)
print(x)
#  \d+ to match the number in a string output formate 50
txt = "The rain in Spain, i love rain 50"
x = re.findall('\d+', txt)
print(x)





```
