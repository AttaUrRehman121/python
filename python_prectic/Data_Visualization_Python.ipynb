```python
#  Tiltle , Labels and Legend 

x_axis = [1,2,3]
y_axis = [4,5,6]

# Plotting line plot with color
plt.title('My First Graph')
plt.plot(x_axis, y_axis, color='red')
```

```python
#  label on x_axis and y_axis 

x_axis = [1,2,3]
y_axis = [4,5,6]
#  label on x_axis and y_axis
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('My First Graph')
plt.plot(x_axis, y_axis, color='red')
```

```python
#  Legend 

x1_axis = [2,4,6,8]
y1_axis = [1,10,100,1000]
x2_axis = [1,3,5,7]
y2_axis = [100,110,120,130]

#  title 

plt.title('Two Lines on One Graph')
plt.xlabel('Amazing X-axis')
plt.ylabel('Incredible Y-axis')
plt.legend(['Line 1', 'Line 2'])

plt.plot(x1_axis, y1_axis, color='red')
plt.plot(x2_axis, y2_axis, color='blue')

```

```python
#  Creating multiple subplots in one figure 
#  ==============================================
#  Creating multiple subplots in one figure (1/2)
#  -----------------------------------------------

x_axis = [2,4,6,8]
y_axis = [1,10,100,1000]
x1_axis = [1,3,5,7]
y1_axis = [100,110,120,130]

plt.subplot(1,2,1)
plt.title('exponential')
plt.xlabel('x_axis')
plt.ylabel('y_axis')
plt.plot(x_axis, y_axis, 'r')

plt.subplot(1,2,2)
plt.title('Linear')
plt.xlabel('x_axis')
plt.ylabel('y_axis')
plt.plot(x1_axis,y1_axis, 'g')

plt.tight_layout()
```
